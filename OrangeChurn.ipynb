## Data preparation

```python
train_df = pd.read_csv('data/churn-bigml-80.csv' )
test_df = pd.read_csv('data/churn-bigml-20.csv' )
```

```python
# Drop redundant charge columns from both train and test datasets
columns_to_drop = ['Total day charge', 'Total eve charge', 'Total night charge', 'Total intl charge']
train_df = train_df.drop(columns=columns_to_drop)
test_df = test_df.drop(columns=columns_to_drop)
```

```python
# Binary encoding for 'International plan' and 'Voice mail plan'
for df in [train_df, test_df]:
    df['International plan'] = df['International plan'].map({'Yes': 1, 'No': 0})
    df['Voice mail plan'] = df['Voice mail plan'].map({'Yes': 1, 'No': 0})

# One-hot encoding for 'State'
train_df = pd.get_dummies(train_df, columns=['State'], drop_first=True)
test_df = pd.get_dummies(test_df, columns=['State'], drop_first=True)

# Ensure both datasets have the same columns
missing_cols = set(train_df.columns) - set(test_df.columns)
for col in missing_cols:
    test_df[col] = 0  
test_df = test_df[train_df.columns]  
```

```python
from sklearn.preprocessing import StandardScaler

# Define numerical columns
numeric_columns = ['Account length', 'Number vmail messages', 'Total day minutes', 'Total eve minutes', 
                   'Total night minutes', 'Total intl minutes', 'Total day calls', 'Total eve calls', 
                   'Total night calls', 'Total intl calls', 'Customer service calls']

# Standardize using StandardScaler
scaler = StandardScaler()
train_df[numeric_columns] = scaler.fit_transform(train_df[numeric_columns])
test_df[numeric_columns] = scaler.transform(test_df[numeric_columns]) 
```

```python
# Separate features and target for training data
X_train = train_df.drop(columns=['Churn'])
y_train = train_df['Churn']

# Separate features and target for testing data
X_test = test_df.drop(columns=['Churn'])
y_test = test_df['Churn']
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to balance the training data
smote = SMOTE(random_state=seed)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)
```

## Model Selection and Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import roc_curve, roc_auc_score
```

#### Logistic regression

```python
# Initialize the Logistic Regression model
log_reg = LogisticRegression(random_state=seed, class_weight='balanced', max_iter=1000)

# Train the model
log_reg.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = log_reg.predict(X_test)
y_pred_prob = log_reg.predict_proba(X_test)[:, 1]  # Probabilities for the positive class
```

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Precision, Recall, F1-Score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
auc_score = roc_auc_score(y_test, y_pred_prob)

plt.figure()
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc_score:.4f})")
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

- #### Obeservations
1- Accuracy (0.7751): The model correctly predicted the outcome for approximately 77.5% of the data. However, this metric can be misleading due to class imbalance.

2- Precision for Churn (0.3632): Only 36.3% of the customers predicted as churners actually churned. This suggests the model has a relatively high false positive rate.

3 -Recall for Churn (0.7684): The model correctly identified 76.8% of the actual churners, which is good for capturing potential churners.

4- F1-Score for Churn (0.4932): Balances precision and recall, showing moderate performance for churn predictions.




So i will try to apply SMOTE (Synthetic Minority Oversampling Technique) to handle the class imbalance and improve the performance of your Logistic Regression model.

```python
# Initialize the Logistic Regression model
log_reg = LogisticRegression(random_state=seed, max_iter=2000)

# Train the model on the resampled data
log_reg.fit(X_train_sm, y_train_sm)
```

```python
# Predict on the test set
y_pred = log_reg.predict(X_test)
y_pred_prob = log_reg.predict_proba(X_test)[:, 1]

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
auc_score = roc_auc_score(y_test, y_pred_prob)

plt.figure()
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc_score:.4f})")
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

#### Random Forset


```python
# Initialize the Random Forest model
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)

# Train the model on the SMOTE-balanced dataset
rf_model.fit(X_train_sm, y_train_sm)

# Make predictions on the test set
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Accuracy:", round(accuracy_rf, 4))

print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
```

```python
# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred_rf)

# Plot confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Churn', 'Churn'], yticklabels=['Non-Churn', 'Churn'])
plt.title("Confusion Matrix - Random Forest with SMOTE")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
# Get ROC curve values
fpr, tpr, thresholds = roc_curve(y_test, rf_model.predict_proba(X_test)[:, 1])

# Plot ROC curve
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='blue', lw=2)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

1-Accuracy (92.95%):The model shows excellent overall accuracy, indicating that the Random Forest is effective in distinguishing between churners and non-churners.

2-Precision for Churn (77%):
The model is reasonably good at correctly predicting the churn class (True). While precision has improved, there is still room for improvement in reducing false positives.

3-Recall for Churn (73%):
Recall shows that the model identifies 73% of churners, which is a solid improvement over earlier models. Still, there's room to increase this percentage to better identify all churners.


4-F1-Score for Churn (75%):
The F1-score is a good balance between precision and recall for churners, showing that the model is performing fairly well in this aspect.


Random Forest has outperformed Logistic Regression, with a notable improvement in the recall and precision for churners.
