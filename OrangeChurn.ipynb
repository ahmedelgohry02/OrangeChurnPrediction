#### Account Length

```python
churn_true = train_df[train_df['Churn'] == True]
churn_false = train_df[train_df['Churn'] == False]
```

```python
# Box plot
sns.boxplot(x='Churn', y='Account length', data=train_df)
plt.title('Account Length vs. Churn')
plt.show()
```

```python
sns.histplot(x = 'Account length' , data = churn_true )
plt.show()
```

```python
sns.histplot(x = 'Account length' , data = churn_false )
plt.show()
```

```python
train_df.groupby('Churn')['Account length'].mean().plot(kind='bar', color=['blue', 'orange'])
plt.title('Mean Account Length by Churn')
plt.ylabel('Average Account Length')
plt.show()
```

- #### Observations

* We can see that account length is normally distrbuted for the both category. And does not have extreme outliers.
* The average account length is nearly the same in this data for th both cayegories.

* Other states, such as MI (Michigan) and AK (Alaska), have no churned users, despite having different sample sizes.
* This could indicate that location-related factors influence customer churn behavior.

2- State Sample Size Imbalance:
* The number of samples from each state varies significantly, ranging from 19 users to 5 users.
* This imbalance may bias state-specific trends or models if not accounted for.

3- Limited Area Codes:

* There are only 3 distinct area codes in the dataset, which might suggest a regional focus or segmentation.
* Area codes could play a role in distinguishing between user groups, but further analysis is needed to evaluate their contribution to churn.

#### Number of calls and charge

```python
tdf = pd.melt(train_df,id_vars='Churn',value_vars=['Total day calls' , 'Total eve calls' , 'Total night calls','Total intl calls'] , var_name = 'call')
# tdf = pd.melt(train_df,id_vars='Churn',value_vars=['Total day charge' , 'Total eve charge' , 'Total night charge'] , var_name = 'call')
sns.barplot(y = 'value' ,x = 'call', data = tdf , hue = 'Churn')
```

```python
tdf = pd.melt(train_df,id_vars='Churn',value_vars=['Total day charge' , 'Total eve charge' , 'Total night charge','Total intl charge'] , var_name = 'Charge')
sns.barplot(y = 'value' ,x = 'Charge', data = tdf , hue = 'Churn')
```

```python
charge_columns = ['Total day charge', 'Total eve charge', 'Total night charge', 'Total intl charge']
churned = train_df[train_df['Churn'] == True][charge_columns].mean()
not_churned = train_df[train_df['Churn'] == False][charge_columns].mean()

print('Churned Customers:\n', churned)
print('Non-Churned Customers:\n', not_churned)
```

- #### Observations

1- Higher Charges for Churned Users:
* The mean charge for users who churned is higher compared to those who didn't.
* This suggests that higher-paying customers might have a higher likelihood of churning.

2- Potential Cause: Pricing Affordability:
* A plausible hypothesis is that Orange's pricing might not align with customer expectations or affordability.
* Customers paying more may feel dissatisfied with the value for money or face financial constraints, leading them to cancel their subscriptions.
